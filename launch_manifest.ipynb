```python
# NOTE: All data used in this script is synthetic and for demonstration purposes only.
# This script focuses on the operational logic of fleet deployment.

import pandas as pd

csv_data = """City,Partner,Compliance,Mapping,Fleet_Ready
Tallinn,Bolt,True,True,True
Helsinki,Alepa,True,False,True
London,UberEats,False,True,False
Stockholm,Wolt,True,True,False
NewYork,DoorDash,True,True,True"""

with open('launch_manifest.csv', 'w') as f:
    f.write(csv_data)

df = pd.read_csv('launch_manifest.csv')
for col in ['Compliance', 'Mapping', 'Fleet_Ready']:
    df[col] = df[col].astype(str).str.strip()

def get_detailed_status(row):
    missing_steps = []

    if row['Compliance'] != 'True':
        missing_steps.append("Legal/Compliance")
    if row['Mapping'] != 'True':
        missing_steps.append("HD Mapping")
    if row['Fleet_Ready'] != 'True':
        missing_steps.append("Fleet Calibration")

    if not missing_steps:
        return "READY"
    else:
        return "MISSING: " + ", ".join(missing_steps)

df['Detailed_Status'] = df.apply(get_detailed_status, axis=1)

print("\n--- 🛰️ DETAILED LAUNCH REPORT() ---")
for index, row in df.iterrows():
    if row['Detailed_Status'] == "READY":
        print(f"✅ {row['City']}: All TASKS completed. Ready to go!")
    else:
        print(f"❌ {row['City']}: {row['Detailed_Status']}")

ready_count = len(df[df['Detailed_Status'] == "READY"])
print("-" * 45)
print(f"TOTAL CITIES READY FOR LAUNCH: {ready_count}")
```
